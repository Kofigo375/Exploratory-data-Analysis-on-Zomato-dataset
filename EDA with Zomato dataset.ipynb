## What do we want to acheive in exploratory data analysis?

2. Explore about numerical variables

```python
df.isnull().sum() # checking the number of missing values with respect to every feature
```

```python
# another way to pull out paritucular column(s) with missing values
# here a list comprehension is used to iterate through all the columns and display the feature(s) with missing values
[features for features in df.columns if df[features].isnull().sum() > 0]
```

```python
df.shape # check the dimension (number of rows and columns in the dataframe)
```

```python
# reading the country code dataset
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# the df and df_country has a common column we can merge them
pd.merge(df, df_country, on='Country Code', how='left')
```

```python
# lets save the merge dataframe as final_df
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.info() # check the colums and datatypes of our final dataframe
```

```python
# another alternative to check only the datatypes is
final_df.dtypes
```

```python
# finding out how many coutries are there
final_df.Country.value_counts()
```

```python
# we can use use the 'index' attribute too store the country names in this order(contries with high number first)
country_names = final_df.Country.value_counts().index
```

```python
# we can also use 'values' attribues to obtain only the values in an array
country_values = final_df.Country.value_counts().values
```

```python
# we can use the country_values and contry_names to plot a pie chart of the top 3 countries that uses zomato 
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observation: Maximum records of Zomato are from India followed by United States and United Kingdom

```python
# group the df by some columns and use '.reset index' method to convert it to a dataframe'
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index() 
```

```python
# rename the last coulumn to 'Rating count' using the 'rename method'
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
# now lets store our new dataframe into a variable 'ratings'
Ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings
```

## Observations
1. when rating is from 4.5 to 4.9 ===> Excellent
2. when rating is from 4.0 to 3.4 ===> Very good
3. when rating is from 3.5 to 3.9 ===> good
4. when rating is from 3.0 to 3.4 ===> average
5. when rating is from 2.5 to 2.9 ===> average
6. when rating is from 2.0 to 2.4 ===> poor
7. Many people gave zero rating

```python
# we can visualize these observations with bar plot
sns.barplot(x="Aggregate rating", y="Rating Count", data=Ratings)
```

```python
# we can use matplot to make the plot bigger and nicer
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)  
sns.barplot(x="Aggregate rating", y="Rating Count", data=Ratings)
```

```python
# we can make the plot more better by using the actual rating colors in the data
sns.barplot(x="Aggregate rating", y="Rating Count",hue='Rating color', data=Ratings, palette=['blue','red','orange','yellow','green','green'])  # added the hue and palette parameter and made it equal to rating color
```

## Observation
    1. Not rated count is very high
    2. Maximum number of rating are between 2.5 and 3.4
